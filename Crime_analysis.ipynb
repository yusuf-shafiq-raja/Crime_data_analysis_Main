```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Crime_Data_2020.csv')
print(df.head())
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Premis Desc'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), cmap='Pastel1', startangle=90)
plt.title("Top 5 Crime Locations")
plt.ylabel("")
plt.show()
```

```python
df['Crm Cd Desc'].value_counts().head(10).plot(kind='bar', figsize=(12, 6), color='lightgreen', edgecolor='black')
plt.title("Top 10 Most Common Crimes")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Vict Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), cmap='coolwarm', startangle=90)
plt.title("Victim Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
df.set_index('DATE OCC').resample('ME').size().plot(kind='line', figsize=(12, 6))
plt.title("Crime Frequency Over Time")
plt.xlabel("Date")
plt.ylabel("Crime Count")
plt.show()
```
